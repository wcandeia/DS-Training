## Estatística: A Ciência das Escolhas

## Introdução

Em um teste [Stroop](https://en.wikipedia.org/wiki/Stroop_effect), participantes são apresentados a uma lista de palavras, e cada uma das palavras está de uma cor. A tarefa dos participantes é dizer em voz alta a cor em que a palavra foi exibida. 
A tarefa tem duas condições: uma condição de palavras congruentes e uma condição de palavras incongruentes. Na condição de palavras congruentes, a palavra exibida são nomes de cores cujos nomes são a mesma cor em que a palavra foi exibida: por exemplo, RED, BLUE. 
Na condição de palavras incongruentes, as palavras apresentadas são nomes de cores cujos nomes não são as mesmas cores em que as palavras foram apresentadas: por exemplo, <font color=purple>PURPLE</font>, <font color=orange>ORANGE</font>. 
Em cada caso, medimos o tempo que demora para o indivíduo falar os nomes em uma lista das cores de tamanhos iguais. Cada participante realizará o teste e tem o tempo medido para a realização do teste nas duas condições.

## Questões para Investigação

**Questão 1**: Qual é a nossa variável independente? Qual é a nossa variável dependente?

**Resposta**: A variável independente é o conjunto de palavras em condição congruente ou incongruente. A variável dependente é o tempo que demora para o indivíduo falar os nomes em uma lista das cores de tamanhos iguais.

**Questão 2a**: Qual seria um conjunto apropriado de hipóteses para essa tarefa? Que tipo de teste estatístico você espera executar? Justifique suas escolhas.

**Resposta**: Existem várias possibilidades: O tempo que demora para o indivíduo falar os nomes em uma lista das cores de tamanhos iguais muda significativamente nas duas condições? O tempo é maior ou menor em uma das condições? Palavras conhecidas pelo participante afetam o desempenho da leitura? O desempenho é significativamente diferente em relação ao gênero do participante?

**Exemplo com notação matemática:**
Hipótese nula: $H_0$: O tempo que demora para o indivíduo falar os nomes em uma lista das cores de tamanhos iguais não muda significativamente nas duas condições. ($\mu_c$ = $\mu_{inc}$, onde: $\mu_c$ é a média de tempo para condição congruente e $\mu_{inc}$ a média de tempo para a condição incongruente.)

Hipótese alternativa: $H_A$: O tempo que demora para o indivíduo falar os nomes em uma lista das cores de tamanhos iguais é maior na condição incongruente. ($\mu_c$ < $\mu_{inc}$, onde: $\mu_c$ é a média de tempo para condição congruente e $\mu_{inc}$ a média de tempo para a condição incongruente.)

A variância da população é desconhecida e é necessário estima-la a partir da amostra. Também temos que o tamanho da amostra é pequeno (n < 30). Assim, o teste T de Student pode ser mais apropriado.

**Questão 2b**: **Resposta**: O tipo das amostras é pareado. Utilizamos teste t pareado (amostras dependentes). O mesmo grupo de indivíduos é testado em duas condições diferentes.

```python
# import all necessary packages and functions.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
sample_data = pd.read_csv('stroopdata.csv') #load data
#display(sample_data.head()) #display data head
#dt_congruent = sample_data['Congruent'].as_matrix()
#dt_incongruent = sample_data['Incongruent'].as_matrix()
dt_congruent = sample_data.Congruent.tolist()
dt_incongruent = sample_data.Incongruent.tolist()

n_data = sample_data.shape[0]

xcong = sample_data['Congruent'].mean() # average 'Congruent'
x_qtiles_cong = sample_data['Congruent'].quantile([.25, .5, .75]).as_matrix() #quantile 'Congruent'

xincong = sample_data['Incongruent'].mean() #average 'Incongruent'
x_qtiles_incong = sample_data['Incongruent'].quantile([.25, .5, .75]).as_matrix() #quantile 'Incongruent'

print('Existem {:d} pontos no conjunto de dados fornecido.\n'.format(n_data*2))
print('A média x da coluna Congruent é cerca de {:.2f}.'.format(xcong))
print('A mediana da coluna Congruent é cerca de {:.2f}.'.format(x_qtiles_cong[1]))
print('O primeiro quartil Q1 é {:.2f}.'.format(x_qtiles_cong[0]))
print('O terceiro quartil Q3 é {:.2f}.\n'.format(x_qtiles_cong[2]))

print('A média x da coluna Incongruent é cerca de {:.2f}.'.format(xincong))
print('A mediana da coluna Incongruent é cerca de {:.2f}.'.format(x_qtiles_incong[1]))
print('O primeiro quartil Q1 é {:.2f}.'.format(x_qtiles_incong[0]))
print('O terceiro quartil Q3 é {:.2f}.\n'.format(x_qtiles_incong[2]))

print('Abaixo um quadro descritivo das principais estatísticas, onde:\ncount = número da amostra para cada condição; mean = média para cada condição; std = desvio padrão da amostra;')
print('min = valor mínimo da amostra; 25% = Q1; 50% = mediana (Q2); 75% = Q3 e max = valor máximo da amostra')

sample_data.describe()
```

**Questão 5**: Agora desempenhe o teste estatístico e reporte seus resultados. Qual seu nível de confiança e o valor estatístico crítico? Você rejeitou a hipótese nula ou falhou ao tentar rejeitá-la? Encontre uma conclusão em relação ao experimento da tarefa. Os resultados estão de acordo com suas expectativas?

**Resposta**: Inicialmente calculamos as médias das amostras, $x_c$ e $x_{inc}$, média para amostra congruente e média para amostra incongruente, respectivamente. Calculamos também a diferença da média: $\mu_c$ - $\mu_{inc}$. Em seguinda calculamos o desvio-padrão para cada amostra: $\sigma_c$ e $\sigma_{inc}$, desvio-padrão para amostra congruente e desvio-padrão para amostra incongruente, respectivamente. O próximo passo foi calcular o desvio-padrão das diferenças $SD$ e o desvio-padrão da média $SEM$. Com essas informações, obtivemos a estatística $t$. Também, consultando a tabela t encontramos o valor $t$-crítico.

Para o intervalo de confiança de $95$%, monocaudal, com $\alpha=0,05$ e grau de liberdade $23$, temos: $IC = (-16,30 ; 0,37)$.

**Questão 6**: Opcional: O que você acha que é responsável pelo efeito observado? Consegue pensar em uma alternativa ou tarefa similar que resultaria em um efeito parecido?

**Resposta**: Estudos indicam que é um fenômeno neurológico relacionado a chamada atenção direta. O efeito Stroop enfatiza a interferência que o processamento automático de palavras tem sobre um maior esforço em dizer apenas o nome da cor, isto é, a tarefa de dar uma resposta adequada quando se tem sinais conflitantes no cérebro (semântica da cor diferente do nome apresentado).

Uma tarefa similar que pode ser testada é utilizar palavras na ordem inversa da escrita ou rotacionadas 180 graus (de cabeça para baixo). Por exemplo: <font color=blue>azul</font> seria apresentado como <font color=blue>luza</font> ou mesmo outra cor, como <font color=red>luza</font>.
